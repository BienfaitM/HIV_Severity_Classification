```python
class Net(nn.Module):
    def __init__(self,input_features=4,hidden1=20,hidden2=20,out_features=2):
        super().__init__()
        self.f_connected1=nn.Linear(input_features,hidden1)
        self.f_connected2=nn.Linear(hidden1,hidden2)
        self.out=nn.Linear(hidden2,out_features)
    def forward(self,x):
        x=F.relu(self.f_connected1(x))
        x=F.relu(self.f_connected2(x))
        x=self.out(x)
        return x

```

```python
model = Net()
```

```python
# optimizer = torch.optim.Adam(model.parameters(), lr = 0.01)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
error = nn.CrossEntropyLoss()
EPOCHS = 500
model.train()

train_losses, test_losses = [], []
for i in range(EPOCHS):
    for features, target in train_loader:
        running_loss = 0
        features = features.float()
        optimizer.zero_grad()
        y_predict = model(features)
        loss = error(y_predict, target)
        loss.backward()
        optimizer.step()
        
        running_loss +=loss.item()
        
    else:
        test_loss = 0
        accuracy  = 0 
        with torch.no_grad():
            model.eval()
            for features, target in test_loader:
                features = features.float()
                y_predict = model(features)
                test_loss = error(y_predict, target)
                
                ps = torch.exp(y_predict)
                top_p, top_class = ps.topk(1, dim=1)
                equals = top_class == target.view(*top_class.shape)
                accuracy = torch.mean(equals.type(torch.FloatTensor))
                
        model.train()
        train_losses.append(running_loss/len(train_loader))
        test_losses.append(running_loss/len(test_loader))
        
        print("Epochs {}/{}..".format(i+1,EPOCHS),
             "training_loss :{:.3f}..".format(train_losses[-1]),
             "test_loss: {:.3f}..".format(test_losses[-1]),
             "Accuracy : {:.3f}..".format(accuracy/len(test_loader)))
                
            
```

```python
class Net2(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(4,40)
        self.fc2 = nn.Linear(40,20)
        self.fc3 = nn.Linear(20,20)
        self.fc4 = nn.Linear(20,10)
        self.fc5 = nn.Linear(10,10)
        self.fc6 = nn.Linear(10,2)
    def forward(self,x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        x = F.relu(self.fc4(x))
        x = F.relu(self.fc5(x))
        x = F.log_softmax(self.fc6(x), dim =1)
        return x
```

```python
model = Net2()
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

epochs = 500
steps = 0

train_losses, test_losses = [], []
for e in range(epochs):
    running_loss = 0
    for images, labels in train_loader:
        
        optimizer.zero_grad()
        images = images.float()
        
        log_ps = model(images)
        loss = criterion(log_ps, labels)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()
        
    else:
        test_loss = 0
        accuracy = 0
        
        # Turn off gradients for validation, saves memory and computations
        with torch.no_grad():
            for images, labels in test_loader:
                images = images.float()
                log_ps = model(images)
                test_loss += criterion(log_ps, labels)
                
                ps = torch.exp(log_ps)
                top_p, top_class = ps.topk(1, dim=1)
                equals = top_class == labels.view(*top_class.shape)
                accuracy += torch.mean(equals.type(torch.FloatTensor))
                
        train_losses.append(running_loss/len(train_loader))
        test_losses.append(test_loss/len(test_loader))

        print("Epoch: {}/{}.. ".format(e+1, epochs),
              "Training Loss: {:.3f}.. ".format(running_loss/len(train_loader)),
              "Test Loss: {:.3f}.. ".format(test_loss/len(test_loader)),
              "Test Accuracy: {:.3f}".format(accuracy/len(test_loader)))
```
