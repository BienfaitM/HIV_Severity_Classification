```python
class Net(nn.Module):
    def __init__(self,input_features=4,hidden1=20,hidden2=20,out_features=2):
        super().__init__()
        self.f_connected1=nn.Linear(input_features,hidden1)
        self.f_connected2=nn.Linear(hidden1,hidden2)
        self.out=nn.Linear(hidden2,out_features)
    def forward(self,x):
        x=F.relu(self.f_connected1(x))
        x=F.relu(self.f_connected2(x))
        x=self.out(x)
        return x

```

```python
model = Net()
```

```python
print(model)
```

```python
for p in model.parameters():
    print(p.shape)
```

```python
optimizer = torch.optim.SGD(model.parameters(),lr=0.001) #lr=0.001
error = nn.CrossEntropyLoss()
EPOCHS = 500
```

```python
# def fit(model, train_loader):
model.train()
final_losses = []

for epoch in range(EPOCHS):
    correct = 0
    for batch_idx, (features, label) in enumerate(train_loader):
        features = features.float()           
        optimizer.zero_grad()
        output = model(features)
        loss = error(output, label)
        loss.backward()
        final_losses.append(loss)
        optimizer.step()
        # Total correct predictions
        predicted = torch.max(output.data, 1)[1] 
        correct += (predicted == label).sum()
        if batch_idx % 50 == 0:
            print('Epoch : {},\t Loss {} \t Accuracy:{:.3f}%'.format(epoch, loss.item(), correct*100/float(BATCH_SIZE*(batch_idx+1))))
```

```python
criterion = nn.CrossEntropyLoss()
otpimizer = torch.optim.Adam(model.parameters(), lr= 0.01)
```

```python
epochs = 500
steps = 0

train_losses, test_losses = [], []
for e in range(epoch):
    for features, target in train_loader:
        running_loss = 0
        features = features.float()
        optimizer.zero_grad()
        log_ps = model(features)
        loss = criterion(log_ps, target)
        loss.backward()
        optimizer.step()
        
        running_loss +=loss.item()
    else:
        test_loss  = 0 
        accuracy   = 0
        with torch.no_grad():
            model.eval()
            for features,target in test_loader:
                features = features.float()
                log_ps = model(features)
                test_loss = criterion(log_ps, target)
                
                ps = torch.exp(log_ps)
                top_p, top_class = ps.topk(1, dim=1)
                equals = top_class == target.view(*top_class.shape)
                accuracy += torch.mean(equals.type(torch.FloatTensor))
        model.train()
        train_losses.append(running_loss/len(train_loader))
        test_losses.append(test_loss/len(test_loader))

        print("Epoch: {}/{}.. ".format(e+1, epochs),
              "Training Loss: {:.3f}.. ".format(train_losses[-1]),
              "Test Loss: {:.3f}.. ".format(test_losses[-1]),
              "Test Accuracy: {:.3f}".format(accuracy/len(test_loader)))
        
        
        
```
