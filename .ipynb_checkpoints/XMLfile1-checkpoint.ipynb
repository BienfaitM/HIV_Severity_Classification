```python
df_cols = ["baselineRNA", "relativeDate", "vLoadMatch", "rawLoad"]
rows = []

for node in xroot.iter("baselineRNA"):
    relativeDate = int(node.find("relativeDate").text)
    vLoadMatch   = node.find("vLoadMatch")
    rawLoad      = int(node.find("rawLoad").text)
    
    rows.append({"relativeDate":relativeDate, "vLoadMatch":relativeDate,"rawLoad":rawLoad})
out_df1 = pd.DataFrame(rows, columns = df_cols)
out_df1
```

```python
def baseline_new_Treatment(out_df3):
    rows = []
    df_col3 = ["baselineNewTreatment","relativeStartDate","relativeStopDate","duration","drug","drugCode","drugClass"]
    for node in xroot.iter("baselineNewTreatment"):
        relativeStartDate = int(node.find("relativeStartDate").text)
        relativeStopDate  = int(node.find("relativeStopDate").text)
        duration          = int(node.find("duration").text)
        for node in xroot.iter("drug"):
            drugCode  = node.find("drugCode").text
            drugClass = node.find("drugClass").text
    rows.append({"relativeStartDate":relativeStartDate,"relativeStopDate":relativeStopDate,
                 "duration":duration,"drugCode":drugCode,"drugClass":drugClass})
    out_df3 = pd.DataFrame(rows, columns = df_col3)
    print(out_df3)
```
