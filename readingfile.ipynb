```python
import pandas as pd
import numpy as np
import os,glob
import xml.etree.ElementTree as ET
```

```python
folder_path = "/Users/mt/Desktop/hivstandford"
filenames = []
for root, directories, files in os.walk(folder_path, topdown=False):
    for name in files:
        filenames.append(os.path.join(root, name)) 
```

```python
roots = []
for index, filename in enumerate(filenames):
    with open(filename, 'r') as f:
        text = f.read()      
        tree = ET.parse(filename)
        root = tree.getroot()
        roots.append(root)
```

```python
df_cols1 = ["baselineRNA", "relativeDate", "vLoadMatch","logLoad","rawLoad"]
rows = []
for root in roots:
    for node in root.iter("baselineRNA"):
        relativeDate = int(node.find("relativeDate").text)
        vLoadMatch   = node.find("vLoadMatch").text
        logLoad      = float(node.find("logLoad").text)
        rawLoad      = int(node.find("rawLoad").text)

        rows.append({"relativeDate":relativeDate, "vLoadMatch":vLoadMatch,"logLoad":logLoad,"rawLoad":rawLoad})
        
out_df1 = pd.DataFrame(rows, columns = df_cols1)
out_df1
```
